```python
recursive_splitter=RecursiveCharacterTextSplitter(
    chunk_size=100,
    chunk_overlap=50,
    length_function=token_length,
    separators=["\n\n", "\n", ".", " "]

)
```

```python
def chunking_laptop_products(all_headers):
    all_chunks=[]
    for doc in all_headers:
        full_description=doc.metadata["Laptop name"] +"\n"+ doc.page_content
        chunk=recursive_splitter.split_text(full_description)
        all_chunks.append(chunk)
    
    return all_chunks
```

```python
from qdrant_client import models
```

```python
def create_collection():
    client.create_collection(
                collection_name=collection_name,
                vectors_config={
                    "dense" : models.VectorParams(
                        size = dense_encoder.embedding_size,
                        distance = models.Distance.COSINE
                    ),
                    "late_interaction" : models.VectorParams(
                        size = late_encoder.embedding_size,
                        distance=models.Distance.COSINE,
                        multivector_config=models.MultiVectorConfig(
                            comparator=models.MultiVectorComparator.MAX_SIM
                        ),
                        hnsw_config=models.HnswConfigDiff(m=0)
                    )
                },  
                sparse_vectors_config= 
                {"sparse":models.SparseVectorParams(modifier=models.Modifier.IDF)}
            )
```

```python
def create_payload():
    client.create_payload_index(
        collection_name=collection_name,
        field_name="product_id",
        field_schema="integer"

    )
    client.create_payload_index(
        collection_name=collection_name,
        field_name="product_name",
        field_schema="keyword"

    )
    client.create_payload_index(
        collection_name=collection_name,
        field_name="url",
        field_schema="keyword"

    )
    client.create_payload_index(
        collection_name=collection_name,
        field_name="chunk_id",
        field_schema="integer"

    )
    client.create_payload_index(
        collection_name=collection_name,
        field_name="chunks",
        field_schema="keyword"

    )
```

```python
def load_to_vectordb():
    with open(file_name,"r",encoding="utf-8") as f:
        markdown_text = f.read()
    headers_to_split_on = [("#","Laptop name")]
    markdown_splitter = MarkdownHeaderTextSplitter(headers_to_split_on=headers_to_split_on)
    all_docs = markdown_splitter.split_text(markdown_text) 
    all_chunks = chunking_laptop_products(all_docs)
    url_pattern =  r"URL:\s*(https://[^\s]+)"

    create_collection()
    create_payload()

    offset = 0
    product_offset = 0
    info = client.get_collection(collection_name=collection_name)
    count = info.points_count
    if(count != 0):
        res,_ = client.scroll(
            collection_name = collection_name,
            limit = 1,
            with_payload = True,
            with_vectors = False,
            order_by={
                "key" : "chunk_id",
                "direction" : "desc"
            }
        )
        if(res):
            product_number = res[0].payload.get("product_id")
            product_offset = product_number + 1
            last_id = res[0].id
            offset = last_id + 1
        else:
            offset = 0
            product_offset = 0
    else:
        offset = 0
        product_offset = 0

    for doc in range(len(all_chunks)):
        for idx in range(len(all_chunks[doc])):
            match = re.search(url_pattern,all_docs[doc].page_content)
            product_url = match.group(1)
            client.upsert(
                collection_name = collection_name,
                points = [
                    models.PointStruct(
                        id = offset,
                        payload = {
                            "product_id" : product_offset,
                            "product_name" : all_docs[doc].metadata["Laptop name"],
                            "url" : product_url,
                            "chunk_id" : offset,
                            "chunk" : all_chunks[doc][idx],
                            
                            
                        },
                        vector = {
                            "dense" : list(dense_encoder.embed(all_chunks[doc][idx]))[0],
                            "sparse" : list(sparse_encoder.embed(all_chunks[doc][idx]))[0].as_object(),
                            "late_interaction" : list(late_encoder.embed(all_chunks[doc][idx]))[0]
                        }
                    )
                ]
            )
            offset+=1
        product_offset+=1
```

```python
load_to_vectordb()
```

```python
import qdrant_client
from typing import List, Dict


def retrieve_chunks(query_text: str, top_k: int = 3) -> List[Dict]:
    
    # 1. Embed the query: Convert the text into a numerical vector
    print(f"Embedding query: '{query_text}'...")
    try:
        query_vector = list(dense_encoder.embed(query_text))[0]
    except Exception as e:
        print(f"Error embedding query: {e}")
        return []

    # 2. Search the Qdrant collection using the 'dense' vector field
    search_results = client.search(
        collection_name=collection_name,
        query_vector=("dense", query_vector), 
        limit=top_k,
        with_payload=True,   
        with_vectors=False  
    )

    # 3. Process the results
    retrieved_data = []
    for hit in search_results:
        retrieved_data.append({
            "score": hit.score,
            "product_name": hit.payload.get("product_name", "N/A"),
            "chunk_text": hit.payload.get("chunk", "N/A"),
           
        })

    return retrieved_data
```

```python
query = "HP OmniBook 7 Aero 13-inch Laptop"
results = retrieve_chunks(query, top_k=3)

print(f"\n--- Search Results for Query: '{query}' ---")

if results:
    for i, res in enumerate(results):
        print(f"[{i+1}] Score: {res['score']:.4f}")
        print(f"    Product: {res['product_name']}")
        print(f"    Chunk: {res['chunk_text']}")
        
        
else:
    print("No results found.")
```
